```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('cleaned_dataset.csv')
```

```python
df['winner_age'] = df.apply(
    lambda row: row['f1_age_when_fight'] if row['winner'] == row['f1'] else row['f2_age_when_fight'],
    axis=1
)

# Count winners by age
age_counts = df['winner_age'].value_counts().sort_index()

# Plot bar graph
plt.figure(figsize=(16, 6))
plt.bar(age_counts.index, age_counts.values)
plt.xlabel('Winner Age')
plt.ylabel('Number of Winners')
plt.title('Number of Winners by Age')
plt.xticks(age_counts.index)
plt.show()
```

```python
# Filter fights where winner's age is over 32
over_32_wins = df[df['winner_age'] > 32]

# Calculate number and percentage of wins by fighters over 32
num_over_32_wins = over_32_wins.shape[0]
total_fights = df.shape[0]
percent_over_32_wins = (num_over_32_wins / total_fights) * 100

print(f"Wins by fighters over age 32: {num_over_32_wins}")
print(f"Percentage of wins by fighters over age 32: {percent_over_32_wins:.2f}%")
```

```python
# Winner distribution
winner_counts = df['winner'].value_counts()
print(f"\nWinner Distribution:")
print(f"Fighter 1 wins: {winner_counts.get(1, 0)} ({winner_counts.get(1, 0)/len(df)*100:.1f}%)")
print(f"Fighter 2 wins: {winner_counts.get(0, 0)} ({winner_counts.get(0, 0)/len(df)*100:.1f}%)")

# Winning methods
winning_method_counts = df['winning_method'].value_counts()
print(f"\nWinning Methods (Top 10):")
for method, count in winning_method_counts.head(10).items():
    percentage = count/len(df)*100
    print(f"  {method}: {count} ({percentage:.1f}%)")

# Visualize winning methods
plt.figure(figsize=(14, 8))

plt.subplot(1, 2, 1)
winning_method_counts.head(10).plot(kind='bar')
plt.title('Top 10 Winning Methods')
plt.xlabel('Winning Method')
plt.ylabel('Number of Fights')
plt.xticks(rotation=45)



plt.tight_layout()
plt.show()
```
